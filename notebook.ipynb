## Visualization & Explanatory Analysis

Memvisualisasikan data dengan menggunakan `matplotlib`

### Pertanyaan 1:

```python
pivot_table_holiday = day_df.pivot_table(index='holiday', values='cnt', aggfunc='sum')
pivot_table_holiday.index = pivot_table_holiday.index.map({0: 'Hari Kerja', 1: 'Hari Libur'})

plt.figure(figsize=(10, 6))
plt.bar(pivot_table_holiday.index, pivot_table_holiday['cnt'], color='skyblue')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Hari Libur')
plt.xlabel('Keterangan Hari')
plt.ylabel('Jumlah Tersewa')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2:

```python
pivot_table_weathersit = day_df.pivot_table(index='weathersit', values='cnt', aggfunc='sum')

weathersit_labels = {
    1: 'Cerah',
    2: 'Berkabut',
    3: 'Cuaca Ringan',
    4: 'Cuaca Ekstrem'
}
pivot_table_weathersit.index = pivot_table_weathersit.index.map(weathersit_labels)

plt.figure(figsize=(10, 6))
plt.bar(pivot_table_weathersit.index, pivot_table_weathersit['cnt'], color='skyblue')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Tersewa')
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

- Conclution pertanyaan 1: Hari Kerja memiliki jumlah penyewaan sepeda yang lebih tinggi dibandingkan dengan Hari Libur, yang menunjukkan bahwa sepeda lebih sering disewa pada hari kerja.
- Conclution pertanyaan 2: Kondisi cuaca mempengaruhi penyewaan, saat cuaca cerah sepeda akan tersewa paling banyak, sedangkan saat cuaca berkabut akan mengalami penurunan penyewaan, cuaca ringan hampir tidak ada penyewaan, dan pada cuaca ekstrem sama sekali tidak ada yang menyewa sepeda
